
# Logistic Regression 

In this project we will be working with a fake advertising data set, indicating whether or not a particular internet user clicked on an Advertisement on a company website. We will try to create a model that will predict whether or not they will click on an ad based off the features of that user.

This data set contains the following features:

* 'Daily Time Spent on Site': consumer time on site in minutes
* 'Age': cutomer age in years
* 'Area Income': Avg. Income of geographical area of consumer
* 'Daily Internet Usage': Avg. minutes a day consumer is on the internet
* 'Ad Topic Line': Headline of the advertisement
* 'City': City of consumer
* 'Male': Whether or not consumer was male
* 'Country': Country of consumer
* 'Timestamp': Time at which consumer clicked on Ad or closed window
* 'Clicked on Ad': 0 or 1 indicated clicking on Ad

```python
#importing libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#reading the data
ad_data = pd.read_csv('advertising.csv')
```

```python
#check the first five values
ad_data.head()
```

```python
ad_data.info()
```

```python
#statastical analysis
ad_data.describe()
```

```python
sns.pairplot(ad_data,hue='Clicked on Ad',palette='bwr')
```

```python
#coeffiient corelation of varibales
plt.figure(figsize=(8,8))
sns.heatmap(ad_data.corr(),annot=True,cmap='Purples')
```

```python
sns.set_style('whitegrid')
ad_data['Age'].hist(bins=30)
plt.xlabel('Age')
```

```python
#separting independent and dependent variable to x and y
X = ad_data[['Daily Time Spent on Site', 'Age', 'Area Income','Daily Internet Usage', 'Male']]
y = ad_data['Clicked on Ad']
```

```python
#splitting x and y data to test and train
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
#fitting the model
logmodel = LogisticRegression()
logmodel.fit(X_train,y_train)
```

## Predictions and Evaluations
** Now predict values for the testing data.**

```python
#prediction on x_test
predictions = logmodel.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
```

```python
#finding accuracy score
print ("Accuracy : ", accuracy_score(y_test, predictions))
```

```python
#confusion matrix
cm = confusion_matrix(y_test, predictions) 
print ("Confusion Matrix : \n", cm)
```

```python
#Create a classification report for the model
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,predictions))
```
